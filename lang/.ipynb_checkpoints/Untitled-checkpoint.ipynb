```python
from collections import defaultdict
import re
import numpy as np

swear_words = ['arse', 'ass',  'bastard', 'bitch', 'bloody',
                'fuck', 'shit', 'damn', 'dick', 'frig', 'hell',
                'jesus', 'piss', 'pussy', 'whore', 'turd', 'twat',
                'wanker']
```

```python
def analyze(community, res, before, after):
    filename = "{}.txt".format(community)
    with open(filename, 'r', encoding = "utf-8") as file:
        for line in file:
            sline = line
            line = line.split(" ")
            for i in range(len(line)):
                for sw in swear_words:
                    if '/' in line[i]:
                        continue
                    s = re.sub(r'[^a-zA-Z0-9]', '', line[i]).lower()
                    if sw in s:
                        res[s] += 1
                        if(i != 0):
                            sbefore = re.sub(r'[^a-zA-Z0-9]', '', line[i-1]).lower()
                            before[s][sbefore] += 1
                        if (i != len(line) - 1):
                            safter = re.sub(r'[^a-zA-Z0-9]', '', line[i+1]).lower()
                            after[s][safter] += 1
```

```python
res_teen = defaultdict(lambda: 0)
before_teen = defaultdict(lambda: defaultdict(lambda: 0))
after_teen = defaultdict(lambda: defaultdict(lambda: 0))
analyze("teenagers", res_teen, before_teen, after_teen)
```

```python
sorted_response_teen = sorted(res_teen.items(), key = lambda x: x[1], reverse = True)
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_raw(community, res_dict, cutoff, color="black", title = "Swear Word Frequency", limit = None):
    sorted_freq = sorted(res_dict.items(), key = lambda x: x[1], reverse = True)
    other = 0
    words = []
    vals = []
    for word, freq in sorted_freq:
        if(freq >= cutoff):
            words.append(word)
            vals.append(freq)
        else:
            other += freq
    words.append("Other")
    vals.append(other)
    fig = plt.figure(figsize = (10,5))
    if(limit != None):
        plt.ylim(0, limit)
    plt.bar(words, vals, color = color, width = 0.4)
    plt.xlabel("r/{}: {}".format(community, title))
    plt.ylabel("Raw Frequency")
    plt.show()
    res = dict()
    for i in range(len(words)):
        res[words[i]] = vals[i]
    return res
```

```python
teen_dict = plot_raw("teenagers", res_teen, 10, "blue")
```

```python
thirty_dict = plot_raw("OVER30REDDIT", res_30, 10, "maroon")
```

```python
def plot_cmp(teen_dict, thirty_dict, title = "Swear Word Comparison"):
    words = []
    teen_vals = []
    teen_other = 0
    thirty_vals = []
    thirty_other = 0
    for word in teen_dict:
        if word in thirty_dict:
            words.append(word)
            teen_vals.append(teen_dict[word])
            thirty_vals.append(thirty_dict[word])
        else:
            teen_other += teen_dict[word]
    for word in thirty_dict:
        if word not in teen_dict:
            thirty_other += thirty_dict[word]

    teen_vals[-1] += teen_other
    thirty_vals[-1] += thirty_other
    X_axis = np.arange(len(words))

    plt.bar(X_axis - 0.2, teen_vals, 0.4, label = 'r/teenagers')
    plt.bar(X_axis + 0.2, thirty_vals, 0.4, label = 'r/OVER30REDDIT')
    plt.xticks(X_axis, words)
    plt.xlabel("Swear Words")
    plt.ylabel("Frequency")
    plt.title("r/teenagers vs r/OVER30REDDIT {}".format(title))
    plt.legend()
    plt.show()
```

```python
plot_cmp(teen_dict, thirty_dict)
```
